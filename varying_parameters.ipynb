```python
from ALMApy import run_alma
import numpy as np
import matplotlib.pyplot as plt
```

### Default parameters

```python
radius = 6371e3
density = 5541
rigidity = 1.46e11
viscosity = 1e21
models = "maxwell"
```

```python
t,k = run_alma(radius, density, rigidity, viscosity, models)
```

### Varying density

```python
t_high_rho,k_high_rho = run_alma(radius, density*2, rigidity, viscosity, models)
t_low_rho,k_low_rho = run_alma(radius, density/2, rigidity, viscosity, models)
```

```python
fig, ax = plt.subplots(figsize=(4,3))
ax.plot(t_low_rho,k_low_rho,label=r"$0.5\rho_\oplus$")
ax.plot(t,k,label=r"$\rho_\oplus$")
ax.plot(t_high_rho,k_high_rho,label=r"$2\rho_\oplus$")
ax.set_xscale("log")
ax.xaxis.set_inverted(True)
ax.set_xlabel(r"$\tau$ (kyr)")
ax.set_ylabel(r"$k_2$")
ax.legend()
fig.tight_layout()
plt.show()
```

### Varying radius

```python
t_low_R,k_low_R = run_alma(radius/2, density, rigidity, viscosity, models)
t_high_R,k_high_R = run_alma(radius*2, density, rigidity, viscosity, models)
```

```python
fig, ax = plt.subplots(figsize=(4,3))
ax.plot(t_low_R,k_low_R,label=r"$0.5R_\oplus$")
ax.plot(t,k,label=r"$1R_\oplus$")
ax.plot(t_high_R,k_high_R,label=r"2$R_\oplus$")
ax.set_xscale("log")
ax.xaxis.set_inverted(True)
ax.set_xlabel(r"$\tau$ (kyr)")
ax.set_ylabel(r"$k_2$")
ax.legend()
fig.tight_layout()
plt.show()
```

### Varying rigidity

```python
t_low_rig, k_low_rig = run_alma(radius, density, rigidity/2, viscosity, models)
t_high_rig, k_high_rig = run_alma(radius, density, rigidity*2, viscosity, models)
```

```python
fig, ax = plt.subplots(figsize=(4,3))
ax.plot(t_low_rig,k_low_rig,label=r"$0.5\mu_\oplus$")
ax.plot(t,k,label=r"$1\mu_\oplus$")
ax.plot(t_high_rig,k_high_rig,label=r"2$\mu_\oplus$")
ax.set_xscale("log")
ax.xaxis.set_inverted(True)
ax.set_xlabel(r"$\tau$ (kyr)")
ax.set_ylabel(r"$k_2$")
ax.legend()
fig.tight_layout()
plt.show()
```

```python
t_low_visc, k_low_visc = run_alma(radius, density, rigidity, viscosity/2, models)
t_high_visc, k_high_visc = run_alma(radius, density, rigidity, viscosity*2, models)
```
